```python
path = '../input.txt'
lines = []

with open(path, 'r') as file:
    for line in file:
        lines.append(line.strip())

for i, line in enumerate(lines):
    lines[i] = list(lines[i])

lines[:10]
```

```python
symbols = []

for x, line in enumerate(lines):
    for y, i in enumerate(line):
        if lines[x][y] != '.' and not lines[x][y].isnumeric():
            symbols.append(lines[x][y])

symbols = list(set(symbols))

symbols[:10]
```

```python
def look_for_numeric(x, y):
    numeric_ids = []

    if x != 0:
        if lines[x - 1][y].isnumeric(): # above
            numeric_ids.append([x - 1, y])
        elif lines[x - 1][y + 1].isnumeric(): # Top right
            numeric_ids.append([x - 1, y + 1])
        elif y != 0:
            if lines[x - 1][y - 1].isnumeric(): # Top left
                numeric_ids.append([x - 1, y - 1])

    if lines[x + 1][y].isnumeric(): # below
        numeric_ids.append([x + 1, y])
    elif lines[x + 1][y + 1].isnumeric(): # bottom right
        numeric_ids.append([x + 1, y + 1])
    elif y != 0:
        if lines[x + 1][y - 1].isnumeric(): # bottom left
            numeric_ids.append([x + 1, y - 1])
        if lines[x][y - 1].isnumeric(): # left
            numeric_ids.append([x, y - 1])

    if lines[x][y + 1].isnumeric(): # right
        numeric_ids.append([x, y + 1])

    return numeric_ids
```

```python
def fetch_number(x, y, max_y):
    current_y = y
    num = ''

    # fetching from right
    while (lines[x][current_y] != '.' and lines[x][current_y] not in symbols) and current_y < max_y - 1:
        num += lines[x][current_y]
        current_y += 1

    # fetching from left
    if y != 0:
        current_y = y - 1

        while (lines[x][current_y] != '.' and lines[x][current_y] not in symbols ) and current_y >= 0:
            num = lines[x][current_y] + num
            current_y -= 1

    return num
```

```python
cords = []

max_y = len(lines[0])

for x, line in enumerate(lines):
    for y, char in enumerate(line):
        if char in symbols:
            for num in look_for_numeric(x, y):
                cords.append(num)

for i in range(0, len(cords)):
    for j in range(i + 1, len(cords)):
        if cords[i][0] == cords[j][0] and cords[i][1] == cords[j][1]:
            del cords[i]

fetched_numbers = []

for cord in cords:
    fetched_numbers.append(int(fetch_number(cord[0], cord[1], max_y)))

print(f'Sum: { sum(fetched_numbers) }\nHow many: { len(fetched_numbers) }')
```
